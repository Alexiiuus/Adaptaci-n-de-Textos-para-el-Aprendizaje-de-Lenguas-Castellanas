```python
#Configs
data_path = "cefr_leveled_texts.csv"
text_column_name = "text"
label_column_name = "level"

model_name = "distilbert-base-uncased"
test_size = 0.25
num_labels = 6  # A1 A2 B1 B2 C1 C2
```

```python
#Read and Prepare Dataset
import pandas as pd
df = pd.read_csv(data_path)
df = df.rename(columns={"label": "level"})
```

```python
#Clean Dataset
df['text_cleaned'] = df["text"]
```

```python
# Cantidad de cada nivel:
df.level.value_counts(dropna=False)
```

```python
#Label Encoder
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
le.fit(df[label_column_name].tolist())
df['label'] = le.transform(df[label_column_name].tolist())
```

```python
# Crear un nuevo dataset con Xcut elementos de cada label:
Xcut = 200
dfcut = pd.concat([df[df['label'] == i].sample(Xcut, random_state=42) for i in range(6)], ignore_index=True)

# Cantidad de cada nivel:
dfcut.level.value_counts(dropna=False)
```
