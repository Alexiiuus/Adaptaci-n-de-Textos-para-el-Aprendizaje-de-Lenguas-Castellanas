# Cohere

```python
import numpy as np

path_adapted_levels = '/CohereFiles/adapted_texts_levels.npy'
path_levels_destined = '/CohereFiles/levels_destined.npy'
adapted_levels = np.load(path_adapted_levels)
levels_destined = np.load(path_levels_destined)
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Generar la matriz de confusión
cm = confusion_matrix(levels_destined, adapted_levels)

# Mostrar la matriz de confusión
plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```
